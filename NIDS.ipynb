```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, roc_curve, auc
from sklearn.ensemble import GradientBoostingClassifier, StackingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from keras.models import Sequential
from keras.layers import Dense, GRU
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import StratifiedKFold, cross_val_score

```

```python
df_dataset = pd.read_csv("network_intrusion.csv")
print("Dataset Loaded: ")
print(df_dataset.head())  
```

```python
class_counts = df_dataset['class'].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='viridis')
plt.title('Class Distribution')
plt.xlabel('Class Labels')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45) 
plt.show()
```

## Data Pre-Processing

### Handling missing values

```python
numeric_cols = df_dataset.select_dtypes(include=['float64', 'int64']).columns
df_dataset[numeric_cols] = df_dataset[numeric_cols].fillna(df_dataset[numeric_cols].mean())

missing_values_found = df_dataset.isnull().sum().any()
print(f"Missing Values Found: {missing_values_found}")
```

### Outlier Detection

```python
numeric_df = df_dataset.select_dtypes(include=['float64', 'int64'])

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

initial_shape = df_dataset.shape
df_dataset = df_dataset[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]
final_shape = df_dataset.shape

print(f"Removed Outliers: {initial_shape[0] - final_shape[0]} rows removed.")
```

```python
# Box plot for numeric columns to visualize outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_df, orient='h')
plt.title('Box Plot for Numeric Columns (Outliers Detection)')
plt.xlabel('Values')
plt.ylabel('Features')
plt.show()
```

### Label Encoding

```python
categorical_cols = df_dataset.select_dtypes(include=['object']).columns

label_encoder = LabelEncoder()
for col in categorical_cols:
    df_dataset[col] = label_encoder.fit_transform(df_dataset[col])

print("Categorical columns encoded.")
```

```python
# Visualize distribution of a few features after encoding
plt.figure(figsize=(14, 8))
for i, col in enumerate(numeric_df.columns[:5], 1):  # Visualizing first 5 numeric features
    plt.subplot(2, 3, i)
    sns.histplot(df_dataset[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Feature Scaling

```python
# Separate features and target
X = df_dataset.drop('class', axis=1)
y = df_dataset['class']

print("Features and target separated.")
print(f"Features shape: {X.shape}, Target shape: {y.shape}")

# Standardization using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

### Splitting the Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Data split into train and test sets: {X_train.shape[0]} train samples, {X_test.shape[0]} test samples.")
```

### Handling Class Imbalance with SMOTE

```python
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"Class imbalance handled. New training samples: {X_train_res.shape[0]}")
```

```python
# Class distribution after SMOTE
plt.figure(figsize=(10, 5))
sns.countplot(x=y_train_res.value_counts().index, palette='viridis')
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation heatmap of numeric features
plt.figure(figsize=(12, 10))
correlation_matrix = df_dataset.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Model Training

## XGBoost

```python
xgb = XGBClassifier()
xgb.fit(X_train_res, y_train_res)
y_pred_xgb = xgb.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb:.4f}")
```

## LightGBM

```python
lgb = LGBMClassifier()
lgb.fit(X_train_res, y_train_res)
y_pred_lgb = lgb.predict(X_test)
accuracy_lgb = accuracy_score(y_test, y_pred_lgb)
print(f"LightGBM Accuracy: {accuracy_lgb:.4f}")
```

## CatBoost

```python
catboost = CatBoostClassifier(verbose=0)
catboost.fit(X_train_res, y_train_res)
y_pred_catboost = catboost.predict(X_test)
accuracy_catboost = accuracy_score(y_test, y_pred_catboost)
print(f"CatBoost Accuracy: {accuracy_catboost:.4f}")
```

## Support Vector Machine (SVM)

```python
svm = SVC()
svm.fit(X_train_res, y_train_res)
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {accuracy_svm:.4f}")
```

## Multi-layer Perceptron (MLP)

```python
mlp = MLPClassifier()
mlp.fit(X_train_res, y_train_res)
y_pred_mlp = mlp.predict(X_test)
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
print(f"MLP Accuracy: {accuracy_mlp:.4f}")
```

## Artificial Neural Network (ANN)

```python
ann = Sequential()
ann.add(Dense(64, input_dim=X_train_res.shape[1], activation='relu'))
ann.add(Dense(32, activation='relu'))
ann.add(Dense(1, activation='sigmoid'))
ann.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

ann.fit(X_train_res, y_train_res, epochs=10, batch_size=32, verbose=0)
accuracy_ann = ann.evaluate(X_test, y_test, verbose=0)[1]
print(f"ANN Accuracy: {accuracy_ann:.4f}")
```

## Gated Recurrent Unit (GRU)

```python
gru = Sequential()
gru.add(GRU(64, input_shape=(X_train_res.shape[1], 1)))
gru.add(Dense(32, activation='relu'))
gru.add(Dense(1, activation='sigmoid'))
gru.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Reshape for GRU input
X_train_res_gru = np.reshape(X_train_res, (X_train_res.shape[0], X_train_res.shape[1], 1))
X_test_gru = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

gru.fit(X_train_res_gru, y_train_res, epochs=10, batch_size=32, verbose=0)
accuracy_gru = gru.evaluate(X_test_gru, y_test, verbose=0)[1]
print(f"GRU Accuracy: {accuracy_gru:.4f}")
```

```python
# Create a dictionary to hold model names and accuracies
model_performance = {
    'XGBoost': accuracy_xgb,
    'LightGBM': accuracy_lgb,
    'CatBoost': accuracy_catboost,
    'SVM': accuracy_svm,
    'MLP': accuracy_mlp,
    'ANN': accuracy_ann,
    'GRU': accuracy_gru
}

# Convert the dictionary into a pandas DataFrame
df_performance = pd.DataFrame(list(model_performance.items()), columns=['Model', 'Accuracy'])

# Print the DataFrame
print(df_performance)
```

## Meta-Model

```python
base_models = [
    ('gbc', GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)),
    ('lgb', LGBMClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)),
    ('xgb', XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)),
    ('mlp', MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=300, random_state=42)),
    ('cat', CatBoostClassifier(iterations=1000, learning_rate=0.1, depth=6, random_seed=42, verbose=0))
]
```

```python
meta_model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

stacking_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    cv=5  
)
```

```python
# Train the stacking model
import joblib
stacking_model.fit(X_train, y_train)

model_filename = 'stacking_model.pkl' 
joblib.dump(stacking_model, model_filename)
```

```python
# Access feature importances from base models
for model in stacking_model.estimators_:
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]

        # Plot feature importances for each base model
        plt.figure(figsize=(10, 7))
        plt.title(f'Feature Importances from {model.__class__.__name__}')
        plt.bar(range(len(importances)), importances[indices], align='center')
        plt.xticks(range(len(importances)), [f'Feature {i+1}' for i in indices], rotation=90)
        plt.xlim([-1, len(importances)])
        plt.show()
```

```python
# Evaluate meta-model performance
stacking_preds = stacking_model.predict(X_test)
accuracy = accuracy_score(y_test, stacking_preds)
cm = confusion_matrix(y_test, stacking_preds)
report = classification_report(y_test, stacking_preds)
```

```python
# Output evaluation results
print(f"Stacking Model Accuracy: {accuracy * 100:.2f}%")
print(f"Confusion Matrix:\n{cm}")
print(f"Classification Report:\n{report}")
```

```python
# Generate classification report
report = classification_report(y_test, stacking_preds, output_dict=True)
report_df = pd.DataFrame(report).transpose()

# Plot precision, recall, and F1-score
plt.figure(figsize=(10, 5))
report_df.iloc[:-1, :-1].plot(kind='bar')
plt.title('Classification Report')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Compute ROC curve and ROC area
fpr, tpr, _ = roc_curve(y_test, stacking_model.predict_proba(X_test)[:, 1], pos_label=1)  # Change `pos_label` as needed
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Perform cross-validation on the stacking model
cv = StratifiedKFold(n_splits=5)
cv_scores = cross_val_score(stacking_model, X_train, y_train, cv=cv, scoring='accuracy')

print(f"Cross-validated Accuracy for Stacking Model: {cv_scores.mean() * 100:.2f}%")
```
